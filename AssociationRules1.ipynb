```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/abhishekgd96/AssociationRules/main/book.csv")
data
```

```python
frequent_books = apriori(data, min_support=0.1, use_colnames=True)
frequent_books
```

```python
rules = association_rules(frequent_books, metric="lift", min_threshold=0.8)
rules
```

```python
rules = association_rules(frequent_books, metric="lift", min_threshold=0.8)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

```python
#An leverage value of 0 indicates independence. Range will be [-1 1]
#A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
rules[['support','confidence']].hist()
```

```python
rules[['support','confidence','lift']].hist()
```

```python
sns.scatterplot('support', 'confidence', data=rules, hue='antecedents')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
